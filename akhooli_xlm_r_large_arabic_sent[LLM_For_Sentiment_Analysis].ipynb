# 5. Load Reviews into DataFrame and Score

```python
import numpy as np
import pandas as pd
df = pd.read_excel('/content/Input Data.xlsx')



# Assuming you have a DataFrame called df and the column name is 'column_name'
df['description'] = df['description'].astype(str)
df
```

```python


# Loop through the DataFrame
for i in range(len(df['description'])):
    # Check if the description is not NaN
    if pd.notna(df['description'].iloc[i]):
        # Tokenize the description
        tokens = tokenizer.encode(df['description'].iloc[i], return_tensors='pt')

        # Get the model's output
        result = model(tokens)

        # Get the predicted class (assuming your model predicts a class)
        predicted_class = int(torch.argmax(result.logits)) + 1

        # Do something with the predicted_class
        print(predicted_class)
    else:
        # Handle NaN values (e.g., skip or set a default value)
        print("Description is NaN at index:", i)
    df['LLM_Result'] = predicted_class
```

```python
print((df['LLM_Result'] == 3).sum())
```

```python
def sentiment_score(review):
    tokens = tokenizer.encode(review, return_tensors='pt')
    result = model(tokens)
    return int(torch.argmax(result.logits))+1
```

```python
import time
df['description'].fillna("just okay", inplace=True)


for i in range(len(df['description'])):
    sentiment_scores = sentiment_score(df['description'].iloc[i])
    print(sentiment_scores)
    df['LLM_Result'] = sentiment_scores
    time.sleep(1)  # Add a 1-second delay
```

```python
print((df['LLM_Result'] == 3).sum())
```
